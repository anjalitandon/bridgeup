```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("nummemdata.csv")
```

```python
df.head(20)
```

```python
unchunk = df["# Correct (UC)"]
chunk = df["# Correct (C)"]
```

```python
len(unchunk)**(1/2)
```

```python
len(chunk)**(1/2)
```

```python
plt.hist(chunk, bins = 4, color = "skyblue")
plt.title("Amount of Correct Numbers, Chunked")
plt.ylabel("Number of Subjects")
plt.xlabel("Amount of Correct Numbers")
plt.savefig("chunk2chunkhist.png")
plt.show()
```

```python
print(unchunk)
```

```python
plt.hist(unchunk, bins = 4, color = "indigo")
plt.title("Amount of Correct Numbers, Unchunked")
plt.ylabel("Number of Subjects")
plt.xlabel("Amount of Correct Numbers")
plt.savefig("chunk2unchunkhist.png")
plt.show()
```

```python
# JUST FOR FUN AND STOLEN FROM MATPLOTLIB BUT I PLAYED AROUND WITH THE DIMENSIONS
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-5, 5, 0.25)
Y = np.arange(-10, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.magma)

plt.show()
```
