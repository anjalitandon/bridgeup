```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("annual_chlorophyll_temp_38N_68W.csv")
```

```python
year = df["Year"]
chlorophyll = df["Chlorophyll (mg/m3)"]
temp = df["Temperature (C)"]
print(year.head())
print(chlorophyll.head())
print(temp.head())
```

```python
plt.clf()
plt.plot(year, chlorophyll, color = 'red')
plt.ylabel("Measurement of Chlorophyll")
plt.legend(loc = 0)
plt.xlabel("Year")
plt.twinx()
plt.plot(year, temp, color = 'blue')
plt.ylabel("Sea Temperature")
plt.xlabel("Year")
plt.xlim(1998, 2017)
plt.legend(loc = 4)
plt.title("Measurement of Chlorophyll \n and Sea Temperature Over Time")
plt.savefig("finalprojectlinegraph.png")
plt.show()
```

```python
import csv
```

```python
csvFile = open("chlorophyll_mgm3_Dec_2002.csv", 'r')
csvReader = csv.reader(csvFile, delimiter = ',')
list_one = list(csvReader)
```

```python
# print(list_one[0])
```

```python
print(len(list_one[0]))
```

```python
print(len(list_one))
```

```python
width = len(list_one[0])
height = len(list_one)
```

```python
grid = np.empty([height, width, 3], dtype=np.uint8)
```

```python
print(grid[0][0])
```

```python
all_1 = []

for r, row in enumerate(list_one):
    for c, value in enumerate(row):
        if float(value) < 99999:
            all_1.append(float(value))
        if float(value) <= 0.05:
            grid[r, c] = [145, 145, 145]
        elif float(value) <= 0.1:
            grid[r, c] = [121, 147, 139]
        elif float(value) <= 0.2:
            grid[r, c] = [145, 188, 175]
        elif float(value) <= 0.3:
            grid[r, c] = [143, 224, 199]
        elif float(value) <= 0.4:
            grid[r, c] = [124, 239, 204]
        elif float(value) <= 0.5:
            grid[r, c] = [89, 249, 201]
        elif float(value) <= 0.75:
            grid[r, c] = [50, 255, 193]
        elif float(value) <= 0.75:
            grid[r, c] = [0, 255, 177]
        elif float(value) <= 15:
            grid[r, c] = [0, 255, 140]
        elif float(value) <= 30:
            grid[r, c] = [0, 255, 100]
        elif float(value) <= 45:
            grid[r, c] = [0, 255, 0]
        elif float(value) <= 60:
            grid[r, c] = [140, 255, 0]
        elif float(value) <= 75:
            grid[r, c] = [191, 255, 0]
        else:
            grid[r, c] = [0, 0, 0]
```

```python
print(grid[100][100])
```

```python
print(max(all_1))
print(min(all_1))
#print(sorted(list(set(all_1))))
```

```python
plt.title("Chlorophyll in Water 2002")
plt.imshow(grid)
plt.axis('off')
plt.savefig("whattheheck2002.png", dpi = 1000)
plt.show()
```
